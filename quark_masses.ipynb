```python
import matplotlib.pyplot as plt
plt.style.use('spectrum.mplstyle')
```

```python
quarks = ["$u$","$d$","$s$"]
masses = [2.2, 4.7, 93]
```

```python
plt.bar(range(3),masses)
plt.ylim(0,95)
plt.yscale("linear")
plt.xticks(range(3),quarks)
plt.ylabel("MeV")
plt.xlabel("quarks")
plt.savefig("quark_mass_chart.pdf")
```
