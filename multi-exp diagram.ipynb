```python
import numpy as np
import h5py
import os, sys
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import image as mpimg
import yaml
import argparse

import utils
import settings
plt.style.use('spectrum.mplstyle')
```

```python
amps = [8.4,5.6,6.6,8.5,9.5]
energies = [0.535,0.645,0.65,0.856,0.977]

def example_corr(n,t):
    ct = np.zeros(len(t))
    for i in range(n):
        ct+=amps[i]*np.exp(-energies[i]*t)
        
    return ct

def contribution(n,t):
    return amps[n]*np.exp(-energies[n]*t)
```

```python
tdata = np.array(list(range(10,40)))
cdata = example_corr(len(amps),tdata)

tfit1 = np.linspace(10,40)
cfit1 = example_corr(2,tfit1)
tfit2 = np.linspace(10,40)
cfit2 = example_corr(3,tfit2)
tfit3 = np.linspace(10,40)
cfit3 = contribution(2,tfit3)
```

```python
nx_plots = 2
ny_plots = 1
figwidth = 16
figheight = 6*ny_plots
label = True

f, axes = plt.subplots(ny_plots, nx_plots, figsize=(figwidth, figheight))

ax = axes[0]
ax.scatter(tdata, cdata)
ax.plot(tfit1,cfit1)
ax.plot(tfit2,cfit2)
ax.plot(tfit3,cfit3)
ax.set_ylabel(r"$C(t)$")
ax.set_xlabel(r"$t/a$")

ax = axes[1]
etdata, edata = utils.effenergy(tdata,cdata)
ax.scatter(etdata, edata)

tfit1,cfit1 = utils.effenergy(tfit1,cfit1)
ax.plot(tfit1,cfit1)
tfit2,cfit2 = utils.effenergy(tfit2,cfit2)
ax.plot(tfit2,cfit2)
tfit3,cfit3 = utils.effenergy(tfit3,cfit3)
# ax.plot(tfit3,cfit3)
ax.set_ylabel(r"$E_{\textup{eff}}(t)$")
ax.set_xlabel(r"$t/a$")
plt.tight_layout()
```
